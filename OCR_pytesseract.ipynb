```python
import cv2
import pytesseract
import matplotlib.pyplot as plt
```

```python
pytesseract.pytesseract.tesseract_cmd='C:\\Program Files\\Tesseract-OCR\\tesseract.exe'
```

```python
image = cv2.imread('alpha.jpg')

plt.rcParams["figure.figsize"]=(16,9)
plt.imshow(image)
```

OpenCV reads images in BGR format and pytesseract understands in RBG format.

We will have to convert from BGR to RBG

```python
img_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(pytesseract.image_to_string(img_rgb))
```

```python
Imgheight, Imgwidth, d  = img_rgb.shape

boxes = pytesseract.image_to_boxes(img_rgb)

for b in boxes.splitlines():
    b=b.split()
    
    x,y,w,h = int(b[1]),int(b[2]),int(b[3]),int(b[4])
    
    cv2.rectangle(img_rgb,(x,Imgheight-y),(w,Imgheight-h),(0,0,255),3)
    cv2.putText(img_rgb,b[0],(x,Imgheight-y),cv2.FONT_HERSHEY_SIMPLEX,1,(255,0,0),2)
    
plt.imshow(img_rgb)
```

```python
image_2 = cv2.imread('testImage_2.png')

plt.rcParams["figure.figsize"]=(16,9)
plt.imshow(image_2)
```

```python
img_2_rgb = cv2.cvtColor(image_2, cv2.COLOR_BGR2RGB)
print(pytesseract.image_to_string(img_2_rgb))
```

```python
hImg, wImg, d  = img_2_rgb.shape
boxes = pytesseract.image_to_data(img_2_rgb)

for x,b in enumerate(boxes.splitlines()):
    if x!=0:
        b=b.split()
    if len(b)==12:
        x,y,w,h = int(b[6]),int(b[7]),int(b[8]),int(b[9])
        cv2.rectangle(img_2_rgb,(x,y),(w+x,h+y),(0,0,255),3)
        cv2.putText(img_2_rgb,b[11],(x,y),cv2.FONT_HERSHEY_COMPLEX,1,(255,0,0),2)
    
plt.imshow(img_2_rgb)
```
